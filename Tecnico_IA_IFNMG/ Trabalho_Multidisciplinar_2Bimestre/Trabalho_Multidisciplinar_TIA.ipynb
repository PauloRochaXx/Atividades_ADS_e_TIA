#### **2.4 QUESTÃO IV** - Os fóruns de discussão são um dos principais espaços de interação entre alunos e professores em cursos a distância. Mas será que participar dos fóruns realmente faz diferença no desempenho?

Crie visualizações de dados, que respondam a pergunta:

*"Alunos que acessam mais os fóruns tiram notas maiores?"*

Antes de criar os gráficos, você precisará realizar a seguinte engenharia de features(engenharia de características):


*   Crie uma nova coluna chamada Total_Forum, somando os
acessos às colunas Fórum Geral, Fórum A, Fórum B e Fórum
C. Essa coluna representa o engajamento total de cada aluno
nos espaços de discussão do curso.

*   Com a nova coluna criada, construa uma visualização de dados que mostre a relação entre os alunos aprovados/reprovados e o engajamento nos fóruns ao longo do curso.



```python
# Criação da coluna "Total_Forum" por meio da soma das participações nos fóruns
df['Total_Forum'] = (
    df['Fórum Geral'] +
    df['Fórum A'] +
    df['Fórum B'] +
    df['Fórum C']
)

# Criar faixas
df['Faixa_Forum'] = pd.cut(
    df['Total_Forum'],
    bins=[-1, 0, 5, 10, 15, 20, float('inf')],
    labels=['0', '1-5', '6-10', '11-15', '16-20', '>20']
)

# Tabela de frequências
tabela = pd.crosstab(df['Faixa_Forum'], df['Resultado'])

# Total de alunos por faixa
totais = tabela.sum(axis=1)

# Percentuais
percentuais = tabela.div(totais, axis=0) * 100

# Gráfico
fig, ax = plt.subplots(figsize=(10, 6))

percentuais[['APROVADO', 'REPROVADO']].plot(
    kind='bar',
    stacked=True,
    color=['steelblue', 'firebrick'],
    ax=ax
)

# Quantidade de alunos no topo
for i, total in enumerate(totais):
    ax.text(
        i,
        102,
        f'{total} alunos',
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Percentuais dentro das barras
for i in range(len(percentuais)):

    aprov = percentuais.iloc[i]['APROVADO']
    reprov = percentuais.iloc[i]['REPROVADO']

    # Parte azul
    ax.text(
        i,
        aprov / 2,
        f'{aprov:.0f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold'
    )

    # Parte vermelha
    ax.text(
        i,
        aprov + reprov / 2,
        f'{reprov:.0f}%',
        ha='center',
        va='center',
        color='white',
        fontweight='bold'
    )

ax.set_title('Participação nos Fóruns x Resultado Final')
ax.set_xlabel('Faixa de Participações nos Fóruns')
ax.set_ylabel('Percentual de Alunos (%)')

ax.set_ylim(0, 110)

plt.legend(title='Resultado')
plt.tight_layout()
plt.show()
```

**2.4.1 - ANÁLISE:** Os dados indicam uma relação positiva entre o engajamento nos fóruns e o resultado final dos estudantes. Observa-se que, à medida que aumenta a participação nos fóruns, cresce a proporção de alunos aprovados no curso. Dessa forma, alunos com maior número de acessos aos fóruns tendem a apresentar melhor desempenho acadêmico.

#### **3.1 QUESTÃO I** - Divida os dados em treino e teste utilizando a razão de 70% para treino e 30% para teste

```python
# Definição da variável alvo (Y), representada pela coluna "Resultado", e das variáveis preditoras (X),
# compostas pelas métricas de acesso e interação dos estudantes no ambiente virtual de aprendizagem
from sklearn.model_selection import train_test_split

y = df['Resultado']
x = df[['Página Geral', 'Fórum Geral', 'Arquivo Conteudo', 'Relatório  ', 'URL Geral', 'Arquivo A', 'Tarefa A', 'URL A', 'Fórum A',
        'Página A', 'Pasta A', 'Questionário A', 'Arquivo B', 'Tarefa B', 'URL B', 'Fórum B', 'Pasta B', 'Questionário B', 'Chat B',
        'Arquivo C', 'Tarefa C', 'URL C', 'Fórum C', 'Questionário C', 'Chat C']]

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)
```

 #### **3.2 QUESTÃO II** - Crie dois modelos de classificação, utilizando a variável Resultado como alvo e todas as variáveis de quantidade de acesso como variáveis descritivas. Não use as variáveis de notas! Cada modelo criado deve utilizar um método diferente de inferência, dentre osmodelos estudados na disciplina.


```python
# Importação dos algoritmos de classificação Random Forest e K-Nearest Neighbors (KNN)
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criação das instâncias dos modelos de classificação
modelo_floresta_radomica = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Ajuste dos modelos aos dados de treinamento
modelo_floresta_radomica.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Avaliação comparativa dos modelos de classificação utilizando a métrica de acurácia
previsao_floresta_radomica = modelo_floresta_radomica.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_floresta_radomica))
display(accuracy_score(y_teste, previsao_knn))
```

**3.2.1 - ANÁLISE:** O modelo Random Forest apresentou acurácia superior à do KNN, indicando maior capacidade de identificar corretamente os alunos aprovados e reprovados a partir das variáveis de acesso ao ambiente virtual.

#### **3.3 QUESTÃO III** - Crie dois modelos de regressão, utilizando a variável NotaTotal como alvo e todas as variáveis de quantidade de acesso como variáveis descritivas. Não use as variáveis de notas! Cada modelo criado deve utilizar um método diferente de inferência, dentre os modelos estudados na disciplina.

```python
# Seleção da variável dependente "NotaTotal" e das variáveis independentes do modelo, seguida da divisão dos
# dados em conjuntos de treinamento e teste
y_nome = ['NotaTotal']
X_nomes = ['Página Geral', 'Fórum Geral', 'Arquivo Conteudo', 'Relatório  ', 'URL Geral', 'Arquivo A', 'Tarefa A', 'URL A', 'Fórum A',
'Página A', 'Pasta A', 'Questionário A', 'Arquivo B', 'Tarefa B', 'URL B', 'Fórum B', 'Pasta B', 'Questionário B', 'Chat B',
'Arquivo C', 'Tarefa C', 'URL C', 'Fórum C', 'Questionário C', 'Chat C']

X = df[X_nomes].values
y = df[y_nome].values

from sklearn.model_selection import train_test_split
X_treinoR, X_testeR, y_treinoR, y_testeR = train_test_split( X, y, test_size=0.3, random_state=0 )
```

```python
# Importação, treinamento e geração de previsões utilizando o algoritmo Árvore de Regressão
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor

modelo_arvore = DecisionTreeRegressor( max_depth = 5 )

modelo_arvore.fit(X_treinoR, y_treinoR)
previsao_arvore = modelo_arvore.predict(X_testeR)

```

```python
# Importação, treinamento e aplicação do modelo de Regressão Linear

from sklearn.linear_model import LinearRegression

modelo_linear = LinearRegression()

modelo_linear.fit(X_treinoR, y_treinoR)
previsao_linear = modelo_linear.predict(X_testeR)
```

```python
# Avaliação do desempenho dos modelos de regressão utilizando as métricas Erro Absoluto Médio (MAE),
# Raiz do Erro Quadrático Médio (RMSE) e Coeficiente de Determinação (R²)

from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score

for nome, previsao in {
    "Linear": previsao_linear,
    "Árvore": previsao_arvore
}.items():

    print(f"\n{nome}")
    print("MAE:", mean_absolute_error(y_testeR, previsao))
    print("RMSE:", root_mean_squared_error(y_testeR, previsao))
    print("R²:", r2_score(y_testeR, previsao))
```

**3.3.1 - ANÁLISE:** A Árvore de Regressão apresentou melhor desempenho que a Regressão Linear, com menor erro (MAE e RMSE) e maior coeficiente de determinação (R²), demonstrando maior capacidade de estimar a nota final dos estudantes.

#### **3.4 QUESTÃO IV** - Faça uma análise de agrupamento dos dados usando um método de redução de dimensionalidade e depois um agrupamento com 3 grupos.

```python
# Aplicação da Análise de Componentes Principais (PCA) para redução da dimensionalidade dos dados,
# seguida da utilização do algoritmo K-Means para agrupamento dos registros

from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

pca = PCA(n_components=3)
X_novo = pca.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=0)
grupos = kmeans.fit_predict(X_novo)
```

```python
# Visualização dos três agrupamentos identificados pelo algoritmo K-Means utilizando
# as duas primeiras componentes principais do PCA

plt.figure(figsize=(8,6))

plt.scatter(
    X_novo[:,0],
    X_novo[:,1],
    c=grupos
)

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Agrupamento dos Alunos')
plt.show()
```
